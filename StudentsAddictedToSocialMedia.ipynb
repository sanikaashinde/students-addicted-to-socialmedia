### Merging

```python
df1 = df[['Student_ID', 'Gender']].copy()
df2 = df[['Student_ID', 'Addicted_Score']].copy()
print(df1)
print(df2)
```

```python
merged_df = pd.merge(df1, df2, on='Student_ID')
print(merged_df.head())
```

### Joins

```python
inner_join = pd.merge(df1, df2, on='Student_ID', how='inner')
print("Inner Join:\n", inner_join.head())
```

```python
left_join = pd.merge(df1, df2, on='Student_ID', how='left')
print("\nLeft Join:\n", left_join.head())
```

```python
right_join = pd.merge(df1, df2, on='Student_ID', how='right')
print("\nRight Join:\n", right_join.head())
```

```python
outer_join = pd.merge(df1, df2, on='Student_ID', how='outer')
print("\nOuter Join:\n", outer_join.head())
```

###  Concatenation 

```python
# Take first 3 rows and next 3 rows
part1 = df.iloc[:3]
part2 = df.iloc[3:6]

# Concatenate rows
concat_rows = pd.concat([part1, part2], axis=0)
print("Row-wise Concatenation:\n", concat_rows)
```

```python
# Use selected columns
col1 = df[['Student_ID']]
col2 = df[['Gender']]
col3 = df[['Addicted_Score']]

# Concatenate columns
concat_columns = pd.concat([col1, col2, col3], axis=1)
print("Column-wise Concatenation:\n", concat_columns.head())
```

### Graphs & Visualizations

```python
plt.figure(figsize=(8,6))
sns.boxplot(x='Affects_Academic_Performance', y='Addicted_Score', data=df)
plt.title("Addiction Score vs Academic Impact")
plt.show()
```

#### Conflicts Over Social Media by Platform

```python
plt.figure(figsize=(12,6))
sns.barplot(x='Most_Used_Platform', y='Conflicts_Over_Social_Media', data=df, ci=None, palette='pastel')
plt.title("Conflicts Over Social Media by Platform")
plt.xticks(rotation=45)
plt.show()
```

### Ranking Students by Social Media Addiction

```python

# Rank students by their addiction score (highest = rank 1)
df['Addiction_Rank'] = df['Addicted_Score'].rank(ascending=False)
df_sorted = df.sort_values(by='Addicted_Score', ascending=False)

print(df_sorted[['Student_ID', 'Addicted_Score', 'Addiction_Rank']].head())
```

### Compare Addiction Scores by Gender

```python
# Group by Gender and find average addiction score
gender_addiction = df.groupby('Gender')['Addicted_Score'].mean()
print(gender_addiction)
```

```python
import matplotlib.pyplot as plt

gender_addiction.plot(kind='bar', color=['orange', 'skyblue'])
plt.title("Average Social Media Addiction by Gender")
plt.ylabel("Average Addiction Score")
plt.xlabel("Gender")
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```

### Compare Addiction by Age Group 

```python
# Bin ages into groups (e.g., 10-15, 16-20)
df['Age_Group'] = pd.cut(df['Age'], bins=[10, 15, 20, 25], labels=['10-15', '16-20', '21-25'])

# Compare addiction scores across age groups
age_addiction = df.groupby('Age_Group')['Addicted_Score'].mean()
print(age_addiction)
```

### Top 5 Most Addicted Students



```python
top_5 = df.sort_values(by='Addicted_Score', ascending=False).head(5)
print(top_5[['Student_ID', 'Gender', 'Addicted_Score']])
```

### Addiction Category Labels

```python
def label_addiction(score):
    if score >= 8:
        return 'High'
    elif score >= 5:
        return 'Moderate'
    else:
        return 'Low'

df['Addiction_Level'] = df['Addicted_Score'].apply(label_addiction)
print(df[['Student_ID', 'Addicted_Score', 'Addiction_Level']].head())
```
